```python
import pandas as pd
import numpy as np

np.random.seed(42)

NB_COLUMN          = 5
NB_ROWS_PRIORITY   = 10
NB_ROWS_DATA       = 30
```

```python
dic_priority = {}
for i in range(NB_COLUMN):
    dic_priority[f'Column_{i}'] = np.random.choice([0, 1, 'X'], size=(NB_ROWS_PRIORITY,), p=[.4, .4, .2])
dic_priority['Output'] = np.random.randint(99999, size=NB_ROWS_PRIORITY)

df_priority = pd.DataFrame(dic_priority)
df_priority.to_csv('data/priority.csv')

df_priority
```

```python
data_row_list = []
for _ in range(NB_ROWS_DATA):
    # Select a random row of priority
    i = np.random.randint(NB_ROWS_PRIORITY)
    row = df_priority.iloc[i]

    # Transform the rwo to remove the X
    for i in range(NB_COLUMN):
        if row[f'Column_{i}'] == 'X':
            row[f'Column_{i}'] = np.random.randint(2)

    # Set the the corresponding line in priority
    data_row_list.append(row)
df_data = pd.DataFrame(data_row_list, columns=df_priority.columns)
df_data.index.names = ['Expected priority correspondance']
df_data.rename({'Output': 'Expected Output'})
df_data.to_csv('data/data.csv')

df_data
```
